#2) Various Types of Help

```python
# List of all indicators
df.ta.indicators()
```

```python
# Help about an indicator such as RSI
help(ta.rsi)
```

#3) RSI

```python
df['RSI_5'] = ta.rsi(\
                   close = df.Close,
                   length = 5)
df.head(20)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize = (42, 5))
plt.plot(df.Date, df.RSI_5)
plt.xticks(rotation=90)
plt.show()
```



---



#4) SMA

```python
df['SMA_5'] = ta.sma(\
                   close = df.Close,
                   length = 5)
df.head(20)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize = (42, 5))

plt.plot(df.Date, df.Close)
plt.plot(df.Date, df.SMA_5)
plt.xticks(rotation=90)
plt.show()
```

#5) Plot Normal Candlestick Chart

#8) ADX

```python
a = ta.adx(df.High,
           df.Low,
           df.Close,
           length = 14)

df = df.join(a)

df.drop(['DMP_14', 'DMN_14'],
        axis = 1,
        inplace = True)

print(df)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize = (42, 5))

#plt.plot(df.Date, df.Close)
plt.plot(df.Date, df.ADX_14)
plt.xticks(rotation=90)
plt.show()
```

#9) Bollinger Bands

```python
import pandas as pd
import pandas_ta as ta

df = pd.read_csv('https://gist.githubusercontent.com/DRALVINANG/5821855d6bcce977fc7f7638bb7ea9a3/raw/9d5bf33a581bf81a8319baf9d677eef309a2d7e9/TSLA%2520Stock%2520Price%2520(2020).csv')

# help(ta.bbands)

bb = ta.bbands(df.Close,
              length = 14,
              std = 2)

bb.drop(['BBB_14_2.0', 'BBP_14_2.0'],
        axis = 1,
        inplace = True)

#renaming columns
bb.columns = ['Lower', 'Mid', 'Upper']

df = df.join(bb)

print(df)

import matplotlib.pyplot as plt

plt.plot(df.Date, df.Close)
plt.plot(df.Date, df.Lower)
plt.plot(df.Date, df.Mid)
plt.plot(df.Date, df.Upper)
plt.xticks(rotation = 90)
plt.show()
```
